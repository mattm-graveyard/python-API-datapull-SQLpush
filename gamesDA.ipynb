```python
import pandas as pd     
import requests
import os
import pyodbc
from sqlalchemy import create_engine
```

```python
url = "https://www.mmobomb.com/api1/games"                                                  #URl used to extract Data made 

response = requests.get(url)                                                                #Requests.get
response
```

```python
responseData = response.json()
#responseData                                                                               #Used for checking while working. Commented out for upload into GitHub for easier readability                              
```

```python
df = pd.json_normalize(responseData)                                                        #Normalizing data for readability 
df                                                                                          #Checking format
```

```python
server = 'MSI'                                                                                                  #Creation for easier use
database = 'master'
tcon = 'Trusted_Connection=True'

conn = pyodbc.connect(Driver='{SQL SERVER}', host = server, database = database, trusted_connection = tcon)     #Connection to SQL Server
cursor = conn.cursor()
```

```python
engine = create_engine('mssql+pyodbc://@MSI/master?trusted_connection=True&driver=ODBC+Driver+17+for+SQL+Server')   #Engine creation 
df.to_sql('Games',engine, if_exists='append',index=True)                                                            #Pushing first version of Table
```

```python
df.columns                                                                                                           #Listing out Columns 
```

```python
columns_to_drop = ['game_url','profile_url','thumbnail','short_description']                                          #Dropping columns I felt were unnecessary for report 
columns_to_drop 
```

```python
df_edited = df.drop(columns=columns_to_drop)                                                                           #Creation of a new DF
df_edited
```

```python
engine = create_engine('mssql+pyodbc://@MSI/master?trusted_connection=True&driver=ODBC+Driver+17+for+SQL+Server')         #Another Engine Instance
df_edited.to_sql('Games_Modified',engine, if_exists='append',index=True)                                                  #Pushing a new table into SSMS 
```
