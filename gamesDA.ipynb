```python
import pandas as pd
import requests
import os
import pyodbc
from sqlalchemy import create_engine
```

```python
url = "https://www.mmobomb.com/api1/games"

response = requests.get(url)
response
```

```python
responseData = response.json()
responseData
```

```python
df = pd.json_normalize(responseData)
df
```

```python
server = 'MSI'
database = 'master'
tcon = 'Trusted_Connection=True'

conn = pyodbc.connect(Driver='{SQL SERVER}', host = server, database = database, trusted_connection = tcon)
cursor = conn.cursor()
```

```python
engine = create_engine('mssql+pyodbc://@MSI/master?trusted_connection=True&driver=ODBC+Driver+17+for+SQL+Server')
df.to_sql('Games',engine, if_exists='append',index=True)
```
